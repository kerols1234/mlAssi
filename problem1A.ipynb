```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('house_data.csv')
```

```python
dataset.insert(4,'ones',1)
```

```python
x = dataset.iloc[:,[4,5]]
y = dataset.iloc[:,2]
```

```python
minValue = x.iloc[:,1].min() 
maxValue = x.iloc[:,1].max()
meanValue = x.iloc[:,1].mean()
```

```python
x.iloc[:,1] = (x.iloc[:,1] - minValue)/(maxValue - minValue)
```

```python
dataSize =  (int)(0.8 * len(x))
trainDataX = x[:dataSize]
testDataX = x[dataSize:]
trainDataY = y[:dataSize]
testDataY = y[dataSize:]
```

```python
arrX = np.array(trainDataX)
arrY = np.array(trainDataY).flatten()
theta = np.array([0,0])
```

```python
m = len(y)
```

```python
def cost_function(x,y,theta,m):
    j = np.sum((x.dot(theta) - y) **2)/(2*m)
    return j
```

```python
iterations = 1000
alpha = 0.005
```

```python
def gradient_descent(x,y,theta,alpha,iterations,m):
    history = [0] * iterations
    for iter in range(iterations):
        hyp = x.dot(theta)
        loss = hyp - y
        gradient = np.sum(x.T.dot(loss))/m
        theta = theta - (alpha * gradient)
        cost = cost_function(x,y,theta,m)
        print(cost)
        history[iter] = cost
    return theta,history
```

```python
(t,h) = gradient_descent(x,y,theta,alpha,iterations,m)
```

```python
#60239358351.00023
plt.title('Cost Function J')
plt.xlabel('No. of iterations')
plt.ylabel('Cost')
plt.plot(h)
plt.show()
```

```python
fitY = [t[0] + t[1]*xx for xx in testDataX.iloc[:,1]]

fig, ax = plt.subplots(figsize=(10,10))
ax.scatter(testDataX.iloc[:,1],testDataY)
ax.plot(testDataX.iloc[:,1],fitY,'-', color='red')

ax.set(title="Simple plot",
      xlabel="x-axis",
      ylabel="y-axis")
```
